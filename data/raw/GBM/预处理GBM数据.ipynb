# 预处理

## 查看共有多少样本，同时具备临床、基因表达、CNV、Mut和甲基化的数据

不考虑突变，共有254个样本同时存在

考虑了突变，仅109个样本。

故，本次试验，多组学数据只采用临床特征、基因表达特征、拷贝数变异特征、甲基化特征

```python
smp_clin = df_clin.attrib_name.to_list()
smp_expr = list(df_expr.columns.values[1:])
smp_mut = list(df_mut.columns.values[1:])
smp_cnv = list(df_cnv.columns.values[1:])
smp_met = list(df_met.columns.values[1:])

print('Clin: {}; Expr: {} ;CNV: {}; Mut: {} ;Met: {}'.format(len(smp_clin), len(smp_expr), len(smp_cnv),len(smp_mut) ,len(smp_met)))
```

```python
smp_tmp = set(smp_clin).intersection(set(smp_expr))
print(len(smp_tmp))

smp_tmp = set(smp_tmp).intersection(set(smp_cnv))
print(len(smp_tmp))

smp_tmp = set(smp_tmp).intersection(set(smp_met))
print(len(smp_tmp))

smp_tmp = set(smp_tmp).intersection(set(smp_mut))
print(len(smp_tmp))
```

```python
# 获取只考虑三种组学的样本名
smp_tmp = set(smp_clin).intersection(set(smp_expr))
smp_tmp = set(smp_tmp).intersection(set(smp_cnv))
smp_tmp = set(smp_tmp).intersection(set(smp_met))
extracted_smp = smp_tmp
```

## 根据样本名提取数据

```python
df_clin_extracted = df_clin[df_clin.attrib_name.isin(extracted_smp)]
df_expr_extracted = df_expr[['attrib_name'] + list(extracted_smp)]
df_cnv_extracted = df_cnv[['attrib_name'] + list(extracted_smp)]
df_met_extracted = df_met[['attrib_name'] + list(extracted_smp)]
```

## 查看LTS和Non-LTS的样本量

```python
df_clin_extracted.status = df_clin_extracted.status.astype('int')
df_clin_extracted.overall_survival = df_clin_extracted.overall_survival.astype('int')
print(df_clin_extracted[(df_clin_extracted.overall_survival < 365*2) & (df_clin_extracted.status == 1)].shape)
print(df_clin_extracted[df_clin_extracted.overall_survival > 365*2].shape)
```

## 查看三种组学的基因重叠情况

```python
print("Expr: {}; CNV: {}; Met: {}".format(df_expr_extracted.shape[0], df_cnv_extracted.shape[0], df_met_extracted.shape[0]))
```

```python
# 同时在三种组学中都存在的基因数

df_cnv_extracted[(df_cnv_extracted.attrib_name.isin(df_met_extracted.attrib_name)) &
                 (df_cnv_extracted.attrib_name.isin(df_expr_extracted.attrib_name))
                ].shape
```

```python
genes = df_cnv_extracted[(df_cnv_extracted.attrib_name.isin(df_met_extracted.attrib_name)) &
                 (df_cnv_extracted.attrib_name.isin(df_expr_extracted.attrib_name))
                ].attrib_name.values
genes
```

```python
# 根据基因名进行数据清洗

df_expr_extracted = df_expr_extracted[df_expr_extracted.attrib_name.isin(genes)]
df_cnv_extracted = df_cnv_extracted[df_cnv_extracted.attrib_name.isin(genes)]
df_met_extracted = df_met_extracted[df_met_extracted.attrib_name.isin(genes)]

print("Expr: {}; CNV: {}; Met: {}".format(df_expr_extracted.shape[0], df_cnv_extracted.shape[0], df_met_extracted.shape[0]))
```

## 基因名转为Entrez-id

```python
import mygene
```

```python
 mg = mygene.MyGeneInfo()
df_tmp = mg.querymany(df_expr_extracted.attrib_name.to_list(), 
                      scopes='symbol', 
                      fields='entrezgene', 
                      species='human', 
                      as_dataframe=True)
df_tmp
```

```python
# 删除异常情况
df_tmp = df_tmp[~df_tmp.entrezgene.isna()]

df_tmp = df_tmp[df_tmp.entrezgene != "7795"]
```

```python
def f(x):
    if x in df_tmp.index.values:
        return df_tmp.loc[x, 'entrezgene']
    else:
        return None
    
df_expr_extracted.attrib_name = df_expr_extracted.attrib_name.map(f)
df_cnv_extracted.attrib_name = df_cnv_extracted.attrib_name.map(f)
df_met_extracted.attrib_name = df_met_extracted.attrib_name.map(f)
```

```python
df_expr_extracted = df_expr_extracted[~df_expr_extracted.attrib_name.isna()]
df_cnv_extracted = df_cnv_extracted[~df_cnv_extracted.attrib_name.isna()]
df_met_extracted = df_met_extracted[~df_met_extracted.attrib_name.isna()]
```

```python
print("Expr: {}; CNV: {}; Met: {}".format(df_expr_extracted.shape[0], df_cnv_extracted.shape[0], df_met_extracted.shape[0]))
```

## 输出

```python
df_expr_extracted.to_csv('../../GBM.expression.tsv.gz', index=False, sep='\t', compression='gzip')
df_cnv_extracted.to_csv('../../GBM.cnv.tsv.gz', index=False, sep='\t', compression='gzip')
df_met_extracted.to_csv('../../GBM.met.tsv.gz', index=False, sep='\t', compression='gzip')
```
